## Handling Missing Values & Outliers

## Data Loading and Initial Inspection

```python
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Fetch data from the API
url = "https://covid-api.com/api/reports?date=2022-04-16&iso=USA"
response = requests.get(url)
data = response.json()

# Convert to DataFrame
df = pd.DataFrame(data['data'])

# Display basic info
print(f"Dataset shape: {df.shape}")
print("\nFirst 5 rows:")
display(df.head())
print("\nData types and missing values:")
display(df.info())
```

## Missing Values Analysis

```python
# Identifying Missing Values

# Check for missing values
missing_values = df.isnull().sum()
print("Missing values per column:")
print(missing_values[missing_values > 0])

# Visual representation of missing values
plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```python

# Strategy for handling missing values:
# 1. Drop columns with excessive missing values
# 2. For numerical columns: impute with median (less sensitive to outliers)
# 3. For categorical columns: impute with mode or create 'Missing' category

# Example implementation (though this dataset may not need it):
cols_with_missing = missing_values[missing_values > 0].index

for col in cols_with_missing:
    if df[col].dtype == 'object':
        # Categorical - impute with mode
        df[col].fillna(df[col].mode()[0], inplace=True)
    else:
        # Numerical - impute with median
        df[col].fillna(df[col].median(), inplace=True)

# Verify no missing values remain
print("\nMissing values after treatment:")
print(df.isnull().sum().sum())
```

## Outliers Detection

```python
# Numerical Variables Analysis
# First, let's see what columns are actually available in the DataFrame
print("Available columns in the DataFrame:")
print(df.columns.tolist())

# Now let's identify which numerical columns are actually present
actual_numeric_cols = []
potential_numeric_cols = ['confirmed', 'deaths', 'recovered', 'active', 'fatality_ratio']

for col in potential_numeric_cols:
    if col in df.columns:
        actual_numeric_cols.append(col)

print("\nNumerical columns available for analysis:")
print(actual_numeric_cols)

# If fatality_ratio isn't present but we want it, we can calculate it
if 'fatality_ratio' not in df.columns and 'confirmed' in df.columns and 'deaths' in df.columns:
    df['fatality_ratio'] = df['deaths'] / df['confirmed']
    actual_numeric_cols.append('fatality_ratio')
    print("\nCreated 'fatality_ratio' column")

# Now proceed with the analysis using only available columns
print("\nDescriptive Statistics:")
display(df[actual_numeric_cols].describe())

# Visualizing distributions and outliers
plt.figure(figsize=(15, 10))
for i, col in enumerate(actual_numeric_cols, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

```python
# Statistical Methods for Outlier Detection

# Using Z-score method (for normally distributed data)
from scipy import stats

def detect_outliers_zscore(data, threshold=3):
    z_scores = np.abs(stats.zscore(data))
    return np.where(z_scores > threshold)

# Using IQR method (more robust)
def detect_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    return ~((data >= (Q1 - 1.5 * IQR)) & (data <= (Q3 + 1.5 * IQR)))

# Detect outliers for each numerical column
outliers_info = {}
for col in numeric_cols:
    outliers = detect_outliers_iqr(df[col])
    outlier_count = outliers.sum()
    outliers_info[col] = {
        'count': outlier_count,
        'percentage': (outlier_count / len(df)) * 100,
        'indices': df.index[outliers]
    }
    print(f"{col}: {outlier_count} outliers ({outliers_info[col]['percentage']:.2f}%)")

# Display outlier information
outliers_df = pd.DataFrame(outliers_info).T
display(outliers_df)
```

```python
#  Handling Outliers

# Strategy selection based on analysis:
# 1. For confirmed cases and deaths - likely real differences between states (not true outliers)
# 2. For fatality ratio - might warrant investigation

# Option 1: Cap outliers (Winsorization)
def winsorize_data(series, lower=0.05, upper=0.95):
    lower_bound = series.quantile(lower)
    upper_bound = series.quantile(upper)
    return series.clip(lower_bound, upper_bound)

# Option 2: Transformations (for skewed data)
def log_transform(series):
    return np.log1p(series)  # log(1+x) to handle zeros

# Option 3: Create indicator variables for outliers
def create_outlier_flags(df, numeric_cols):
    for col in numeric_cols:
        df[f'{col}_outlier'] = detect_outliers_iqr(df[col])
    return df

# Implementation example:
# We'll winsorize fatality ratio but keep original values for cases/deaths
df['fatality_ratio_winsorized'] = winsorize_data(df['fatality_ratio'])

# Compare original vs winsorized
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.boxplot(y=df['fatality_ratio'])
plt.title('Original Fatality Ratio')
plt.subplot(1,2,2)
sns.boxplot(y=df['fatality_ratio_winsorized'])
plt.title('Winsorized Fatality Ratio')
plt.tight_layout()
plt.show()
```

## Cumulative Distribution Plot

```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Step 1: Fetch global COVID-19 data
url = "https://covid-api.com/api/reports?date=2022-04-16"
response = requests.get(url)
data = response.json()

# Step 2: Process data
df = pd.json_normalize(data['data'])
df = df[df['region.province'].isna()]  # Country-level data only

# Continent mapping (simplified - use pycountry_convert for full mapping)
continent_map = {
    'USA': 'North America', 'CAN': 'North America',
    'BRA': 'South America', 'ARG': 'South America',
    'GBR': 'Europe', 'FRA': 'Europe', 'DEU': 'Europe',
    'CHN': 'Asia', 'IND': 'Asia', 'JPN': 'Asia',
    'ZAF': 'Africa', 'NGA': 'Africa', 'EGY': 'Africa',
    'AUS': 'Oceania', 'NZL': 'Oceania'
}
df['continent'] = df['region.iso'].map(continent_map)
df = df.dropna(subset=['continent'])

# Step 3: Create CDF plots
metrics = ['confirmed', 'deaths', 'recovered', 'active']
plt.figure(figsize=(15, 10))

for i, metric in enumerate(metrics, 1):
    plt.subplot(2, 2, i)

    for continent in df['continent'].unique():
        subset = df[df['continent'] == continent][metric]
        if len(subset) > 0:
            # Sort values and calculate cumulative probability
            sorted_values = np.sort(subset)
            cum_prob = np.arange(1, len(sorted_values)+1) / len(sorted_values)

            # Plot CDF
            plt.plot(sorted_values, cum_prob, label=continent, linewidth=2)

    plt.title(f'CDF of {metric.capitalize()}')
    plt.xlabel(metric.capitalize())
    plt.ylabel('Cumulative Probability')
    plt.grid(True, alpha=0.3)
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import mstats
import numpy as np

# Check if column exists and has valid data
if 'deaths' in df.columns:
    deaths_data = df['deaths'].dropna()  # Remove missing values

    if len(deaths_data) > 0:  # Only proceed if we have data
        try:
            # Winsorize only if we have enough data points
            if len(deaths_data) >= 10:  # Minimum threshold
                df['winsorized_deaths'] = mstats.winsorize(
                    deaths_data,
                    limits=[0.05, 0.05]
                )
            else:
                print("Not enough data points for winsorization (n < 10)")
                df['winsorized_deaths'] = deaths_data
        except Exception as e:
            print(f"Winsorization failed: {e}")
            df['winsorized_deaths'] = deaths_data
    else:
        print("No valid death data available")
        df['winsorized_deaths'] = np.nan
else:
    print("'deaths' column not found in DataFrame")
```

### Spearman Rank Correlation (Numerical vs Categorical/Ordinal)

```python
import requests
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# Fetch and prepare data
url = "https://covid-api.com/api/reports?date=2022-04-16&iso=USA"
response = requests.get(url)
data = response.json()
df = pd.DataFrame(data['data'])

# First, let's see what columns we actually have
print("Available columns in the dataset:")
print(df.columns.tolist())

# Create categorical variables based on available columns
# For US data, we can use 'region.iso' (state codes) or 'region.name' (state names) instead of province
df['region_type'] = 'State'  # All entries are states/territories in US data

# Create severity categories
df['cases_severity'] = pd.cut(df['confirmed'],
                             bins=[0, 500000, 1000000, np.inf],
                             labels=['Low', 'Medium', 'High'])

# Calculate fatality ratio if not present
if 'fatality_ratio' not in df.columns and 'deaths' in df.columns and 'confirmed' in df.columns:
    df['fatality_ratio'] = df['deaths'] / df['confirmed']

# Select numerical columns that actually exist
numerical_cols = []
for col in ['confirmed', 'deaths', 'active', 'fatality_ratio']:
    if col in df.columns:
        numerical_cols.append(col)

# Calculate Spearman correlation matrix
spearman_corr = df[numerical_cols].corr(method='spearman')

print("\nSpearman Rank Correlation Matrix:")
display(spearman_corr)

# Visualize
plt.figure(figsize=(10,8))
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', center=0, vmin=-1, vmax=1)
plt.title('Spearman Rank Correlation Matrix')
plt.show()

# Between Numerical and Ordinal Variables
if 'cases_severity' in df.columns:
    df['severity_rank'] = df['cases_severity'].cat.codes
    if 'confirmed' in df.columns:
        rho, pval = stats.spearmanr(df['confirmed'], df['severity_rank'])
        print(f"\nSpearman correlation between confirmed cases and severity rank: {rho:.3f} (p={pval:.4f})")
    else:
        print("\n'confirmed' column not available for severity correlation")
else:
    print("\nCould not create severity categories for correlation analysis")

# Additional visualization: Boxplot of deaths by severity
if 'cases_severity' in df.columns and 'deaths' in df.columns:
    plt.figure(figsize=(10,6))
    sns.boxplot(data=df, x='cases_severity', y='deaths')
    plt.title('COVID-19 Deaths by Case Severity Category')
    plt.show()
```

## Enhanced Correlation Heatmap

```python
import numpy as np

# Calculate correlation matrix
corr_matrix = df[numerical_vars].corr()

# Check for and handle infinite/missing values
print("Missing values before handling:", corr_matrix.isnull().sum().sum())
print("Infinite values before handling:", np.isinf(corr_matrix.values).sum())

# Fill any remaining NA/NaN values with 0 (no correlation)
corr_matrix = corr_matrix.fillna(0)

# Replace infinite values with reasonable bounds
corr_matrix = corr_matrix.replace([np.inf, -np.inf], 0)

# Verify the matrix is clean
print("\nMissing values after handling:", corr_matrix.isnull().sum().sum())
print("Infinite values after handling:", np.isinf(corr_matrix.values).sum())

# Create clustered heatmap with error handling
try:
    plt.figure(figsize=(10, 8))
    g = sns.clustermap(corr_matrix,
                      annot=True,
                      cmap='coolwarm',
                      center=0,
                      vmin=-1,
                      vmax=1,
                      figsize=(10, 8),
                      dendrogram_ratio=0.2,
                      cbar_pos=(0.02, 0.8, 0.05, 0.18))
    plt.title('Clustered Correlation Matrix of COVID-19 Metrics', y=1.1)

    # Adjust layout to prevent title cutoff
    g.fig.subplots_adjust(top=0.9)
    plt.show()

except Exception as e:
    print(f"Error creating clustermap: {e}")
    print("Creating standard heatmap instead...")

    plt.figure(figsize=(8, 6))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, vmin=-1, vmax=1)
    plt.title('Correlation Matrix of COVID-19 Metrics')
    plt.show()
```

###	Test Statistics  
###  Develop a hypothesis and select the appropriate test statistics based on your data:
Test Selection: Pearson correlation test or linear regression

```python
# Hypothesis: States with higher case counts have significantly different fatality rates

from scipy import stats

# Calculate Pearson correlation and p-value
corr, p_value = stats.pearsonr(df['confirmed'], df['fatality_ratio'])
print(f"Pearson Correlation: {corr:.3f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis - Significant correlation exists")
else:
    print("Fail to reject null hypothesis - No significant correlation")
```

```python
# Hypothesis: Fatality ratios differ between high and low case count states

# Split into groups
high_case = df[df['cases_severity'] == 'High']['fatality_ratio']
low_case = df[df['cases_severity'] == 'Low']['fatality_ratio']

# Check normality
_, p_high = stats.shapiro(high_case)
_, p_low = stats.shapiro(low_case)

if p_high > 0.05 and p_low > 0.05:
    # Parametric t-test
    t_stat, p_val = stats.ttest_ind(high_case, low_case)
    print(f"\nT-test Results: t={t_stat:.3f}, p={p_val:.4f}")
else:
    # Non-parametric Mann-Whitney
    u_stat, p_val = stats.mannwhitneyu(high_case, low_case)
    print(f"\nMann-Whitney U Results: U={u_stat:.3f}, p={p_val:.4f}")
```
