""" This EDA examines COVID-19 data from the United States on April 16, 2022, obtained from the COVID-API. The analysis will assess data adequacy and integrity, explore patterns, and identify relationships between variables."""

## Data Loading and Initial Inspection

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Fetch data from the API
url = "https://covid-api.com/api/reports?date=2022-04-16&iso=USA"
response = requests.get(url)
data = response.json()

# Convert to DataFrame
df = pd.DataFrame(data['data'])

# Display basic info
print(f"Dataset shape: {df.shape}")
print("\nFirst 10 rows:")
display(df.head())
print("\nData types and missing values:")
display(df.info())
```

"""  Dataset Size and Representativeness
Size: The dataset contains 56 entries (one for each US state/territory)

Representativeness: Covers all 50 states plus territories like Puerto Rico and Guam

Variables: Includes confirmed cases, deaths, fatality ratio, and regional breakdowns

Relevant Variables Analysis
Key variables present:

confirmed: Total confirmed cases

deaths: Total deaths

fatality_ratio: Deaths per confirmed cases

region: State/territory information

active: Currently active cases

recovered: Recovered cases

The dataset appears adequate for analyzing COVID-19 patterns across US regions."""

## Data Cleaning and Preparation

```python
# Handle missing values if any
# For this dataset, there are no missing values in key columns

# Convert date to datetime format
df['date'] = pd.to_datetime(df['date'])

# Calculate additional metrics
df['mortality_rate'] = (df['deaths'] / df['confirmed']) * 100
df['recovery_rate'] = (df['recovered'] / df['confirmed']) * 100
df['active_rate'] = (df['active'] / df['confirmed']) * 100
```

## Exploratory Data Analysis

```python
# Univariate Analysis

# Check if fatality_ratio exists, if not create it
if 'fatality_ratio' not in df.columns and all(col in df.columns for col in ['deaths', 'confirmed']):
    df['fatality_ratio'] = df['deaths'] / df['confirmed']

# Adjust subplots based on available columns
available_columns = [col for col in ['confirmed', 'deaths', 'fatality_ratio', 'active'] if col in df.columns]

fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes = axes.ravel()  # Flatten the axes array for easier iteration

for i, col in enumerate(available_columns[:4]):  # Show up to 4 plots
    sns.histplot(df[col], bins=20, ax=axes[i], kde=True)
    axes[i].set_title(f'Distribution of {col.replace("_", " ").title()}')

# Hide any unused subplots
for j in range(len(available_columns), 4):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
```

## Multivariate Analysis

```python
# First, identify numeric columns (you can select specific ones or automatically detect them)
# Option 1: Select specific numeric columns you're interested in
# numeric_cols = ['cases', 'deaths', 'recoveries', ...]

# Option 2: Automatically select all numeric columns
numeric_cols = df.select_dtypes(include=['number']).columns.tolist()

# Then proceed with your correlation matrix
corr_matrix = df[numeric_cols].corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of COVID-19 Metrics')
plt.show()

# Pairplot of key variables
sns.pairplot(df[numeric_cols])
plt.show()
```
